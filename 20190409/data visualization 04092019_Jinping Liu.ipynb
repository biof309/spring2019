```python
#multi-data scatter plot
# evenly sampled time at 0.2 intervals
#np.arrage
t = np.arange(0., 5., 0.2)
t1 = t**2
t2 = t**3
# red dashes, blue squares and green triangles
#label = '', plt.legend(), loc = 'upper left'
plt.plot(t, t, 'r--', label= 't')
plt.plot( t, t1, 'bs',label ='t1')
plt.plot( t, t2, 'g^', label = 't2')
plt.legend(loc = 'upper left')
plt.show()
```

```python
#scatter plot with size and color__bubble plot
#np.arange
#np.random.randint
#np.random.randn
#np.abs
data = {'a': np.arange(50), # Return evenly spaced values within a given interval. 
        'c': np.random.randint(0, 50, 50), # Return integers
        'd': np.random.randn(50)} # Return a sample (or samples) from the standard normal distribution
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data, alpha=0.6) # alpha for transparency
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.colorbar();  # show color scale
plt.show()
```

### Histograms, Binnings, and Density

```python
plt.style.use('seaborn-white')
data = np.random.randn(2000)  
```

```python
plt.hist(data) #plt.hist
plt.show()
```

```python
plt.hist(data, 
         bins=30, 
         density=True, 
         alpha=0.5, 
         color='steelblue',
         edgecolor='none')
plt.show()
```

```python
#np.random.normal()
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(bins=40,
              histtype='stepfilled', 
              alpha=0.3, 
              density=True)

#using *args and **kwargs to call a function
#Motif: *args and **kwargs serves as a placeholder for the arguments 
#that need to be passed to a function call
plt.hist(x1, **kwargs) 
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
plt.show()
```

```python
#density plot using import seaborn as sns

p1=sns.kdeplot(x1, shade=True, label = 'x1')
p1=sns.kdeplot(x2, shade=True,label = 'x2')
p1=sns.kdeplot(x3, shade=True, label = 'x3')
plt.legend(loc = 'upper left')
plt.show()
```

### Boxplot

```python
import seaborn as sns

# Load the example tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
#Boxplot with jitter
ax = sns.boxplot(x="day", y="total_bill", data=tips)
ax = sns.stripplot(x="day", y="total_bill", data=tips, color="orange", jitter=0.2, size=2.5)
plt.title("Boxplot with jitter", loc="left")
```

```python
ax = sns.violinplot(x="day", y="total_bill", data=tips)
ax = sns.stripplot(x="day", y="total_bill", data=tips, color="darkblue", jitter=0.2, size=2.5)
```

### Nested violin plot

```python
# set background
sns.set(style="whitegrid", palette="pastel", color_codes=True)

# days versus total bill
sns.violinplot(x="day", y="total_bill", data=tips)
sns.despine(left=True)
plt.legend(loc = 'upper left')
```

```python
# set background
sns.set(style="whitegrid", palette="pastel", color_codes=True)

# days versus total bill
sns.violinplot(x="day", y="total_bill", hue="smoker",
               split=True, inner="quart",
               palette={"Yes": "y", "No": "b"},
               data=tips)  # palette=["lightblue", "lightpink"])
sns.despine(left=True)
plt.legend(loc = 'upper left')
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",
               split=True, inner="quart",
               palette={"Yes": "lightblue", "No": "lightpink"} ,
               data=tips)  # palette=["lightblue", "lightpink"])
sns.despine(left=True)
plt.legend(loc = 'upper left')
```

### Heatmap

```python
# Data set# Heatmap
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt
 
url = 'https://python-graph-gallery.com/wp-content/uploads/mtcars.csv'
df = pd.read_csv(url)
df = df.set_index('model')
df
 
```

```python
sns.clustermap(df)
```

### scaling data

```python
# Standardize or Normalize every column in the figure
# Standardize:
sns.clustermap(df, standard_scale=1) # subtract the minimum and divide each by its maximum
# Normalize
sns.clustermap(df, z_score=1)  # z = (x - mean)/std
```

```python
# OK now we can compare our individuals. But how do you determine the similarity between 2 cars?
# Several way to calculate that. the 2 most common ways are: correlation and euclidean distance?
sns.clustermap(df, metric="correlation", standard_scale=1)
sns.clustermap(df, metric="euclidean", standard_scale=1)
```

### Method to determine distance link between clusters

```python
# OK now we determined the link distance link between 2 clusters. But how to do the clusterisation? Several methods exist.
# If you have no idea, ward is probably a good start.
#complete, average, centroids, and ward links
sns.clustermap(df, metric="euclidean", standard_scale=1, method="single") #closet elements between clusters
sns.clustermap(df, metric="euclidean", standard_scale=1, method="ward") # aggregate diviation 
```

### Change color palette

```python
# Change color palette
sns.clustermap(df, metric="euclidean", standard_scale=1, method="ward", cmap="mako")
sns.clustermap(df, metric="euclidean", standard_scale=1, method="ward", cmap="viridis")
sns.clustermap(df, metric="euclidean", standard_scale=1, method="ward", cmap="Blues")
```

### side bar

```python
# Prepare a vector of color mapped to the 'cyl' column
row_colors = df.cyl.map(my_palette)
 
# plot
sns.clustermap(df, metric="correlation", method="single", cmap="Blues", standard_scale=1, row_colors=row_colors)
```
